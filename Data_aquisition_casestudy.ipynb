#Step 1: Load SpaceX Launch Data from API

```python
import numpy as np
import pandas as pd
import sqlite3
import requests
```

Load

```python
url = "https://api.spacexdata.com/v4/launches"
data = requests.get(url).json()
```

```python
launch_df=pd.DataFrame(data)
launch_df.head(10)
```

```python
launch_df.info()
```

Extract

```python
launch_df=launch_df[["name", "date_utc", "success", "details", "rocket"]]
launch_df.head()
```

Convert

```python
launch_df["date_utc"] = pd.to_datetime(launch_df["date_utc"])
launch_df["year"] = launch_df["date_utc"].astype(str).str.split("-").str[0]
launch_df[["date_utc", "year"]]
```

#Step 2: Load Rocket Metadata

Load

```python
rocket_df=pd.DataFrame(requests.get("https://api.spacexdata.com/v4/rockets").json())
rocket_df.head(10)
```

```python
rocket_df.info()
```

Extract

```python
rocket_df=rocket_df[["id", "name", "type", "active", "stages"]]
rocket_df.head()
```

#Step 3: Merge Launch and Rocket Data

```python
launch_data=launch_df.rename(columns={'name': 'launch_name'})
rocket_data=rocket_df.rename(columns={'name': 'rocket_name'})


merged_df = pd.merge(
    launch_data,
    rocket_data,
    left_on='rocket',
    right_on='id',
    how='left'
)
merged_df.head(10)
```

```python
merged_df.info()
```

#Step 4: Add Simulated Country Information

```python
countries = ['USA', 'Russia', 'India', 'China', 'France']
merged_df['country']=np.random.choice(countries,size=len(merged_df))

merged_df.head()
```

#Step 5: Store Merged Data in SQLite3

```python
conn=sqlite3.connect('launches.db')
```

```python
merged_df.to_sql('launches', conn,index=False)
```

```python
df_db = pd.read_sql('SELECT * FROM launches LIMIT 5', conn)
df_db
```

#Step 6: Run SQL Queries on the Data to analyze


Launches by Country

```python
cursor=conn.cursor()
```

```python
cursor.execute("""
     SELECT country, COUNT(*) AS launches_count
    FROM launches
    GROUP BY country
    ORDER BY launches_count DESC;
    """
)
```

```python
cursor.fetchall()
```

Which year had the highest number of launches?

```python
cursor.execute("""
    SELECT year, COUNT(*) AS launches_count
    FROM launches
    GROUP BY year
    ORDER BY launches_count DESC
    LIMIT 1;
""")
```

```python
cursor.fetchone()
```

Top 5 Missions by Launch Count

```python
cursor.execute("""
    SELECT launch_name, COUNT(*) AS launch_count
    FROM launches
    GROUP BY launch_name
    ORDER BY launch_count DESC
    LIMIT 5;
""")
```

```python
cursor.fetchall()
```

Top 3 by rocket count

```python
cursor.execute("""
    SELECT rocket_name, COUNT(*) AS launch_count
    FROM launches
    GROUP BY rocket_name
    ORDER BY launch_count DESC
    LIMIT 5;
""")
```

```python
cursor.fetchall()
```
