#task 8:Handle missing values in 'Revenue (Millions)' and 'Metascore

```python
df.info()   #finding null values
```

#task 16:  Create a new column Profitability:
a) Revenue divided by Duration
b) Find top 10 most profitable movies

```python
df['Profitability'] = df['Revenue (Millions)'] / df['Runtime (Minutes)']
top_10_profit = df.sort_values(by='Profitability', ascending=False).head(10)
print(top_10_profit[['Title', 'Revenue (Millions)', 'Runtime (Minutes)', 'Profitability']])
```

#task 17 :Create a column that categorizes movies by rating:
a) Rating < 6: Poor
b) 6 ≤ Rating < 7.5: Average
c) Rating ≥ 7.5: Excellent

```python
conditions = [ df['Rating'] < 6,
    (df['Rating'] >= 6) & (df['Rating'] < 7.5),
    df['Rating'] >= 7.5]


values = ['Poor', 'Average', 'Excellent']

df['Rating_Category'] = np.select(conditions, values, default='Unknown')

print(df[['Title', 'Rating', 'Rating_Category']])
```

#task 18:Create a custom function to score movies based on:
i)
Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
Apply a function to dataset to assign the score

```python
def calculate_high_score(row):
    return (row['Rating'] * 10) + row['Revenue (Millions)']


df['High_Score'] = df.apply(calculate_high_score, axis=1)


scored_movies = df.sort_values(by='High_Score', ascending=False)

print(scored_movies[['Title', 'Rating', 'Revenue (Millions)', 'High_Score']])
```
