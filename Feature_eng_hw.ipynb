```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris.info()
```

```python
plt.plot(iris.sepal_length,iris.sepal_width,color='blue',marker='o',linestyle='dashed',markerfacecolor='red',markersize=10)
```

```python
plt.plot(iris.sepal_length,iris.petal_length)
avg_length=np.mean(iris.petal_length)
plt.axhline(y=avg_length,color='r',linestyle='--')
plt.xlabel('sepal length')
plt.ylabel('petal length')
plt.title('sepal vs petal')
plt.legend(['sepal length','avg petal length'])
```

```python
plt.hist(iris.petal_width,bins=10,edgecolor='red')
plt.show()
```

```python
fg,ax=plt.subplots(2,2,figsize=(10,4))
ax[0,0].plot(iris.petal_length,iris.petal_width)
ax[0,1].scatter(iris.petal_length,iris.petal_length)
ax[1,0].bar(iris.sepal_length,iris.sepal_width)
ax[1,1].hist(iris.petal_width,bins=10,edgecolor='red')
plt.show()
```

```python
plt.pie(iris.petal_length.values,labels=iris.species)
plt.show()
```

```python
plt.pie(iris.species.value_counts(),labels=iris.species.value_counts().index)
plt.show()
```

```python
sns.boxplot(data=iris)
```

```python
sns.histplot(iris,kde=True)
```

```python
sns.kdeplot(iris)
```

```python
#  relplot - Scatter plot with facets
sns.relplot(
    x='petal_length', y='petal_width',
    hue='species', col='species', # Facet by species
    kind='scatter', data=iris,
    height=4
).fig.suptitle(' Relational Plot (Faceting by Species)', y=1.02)
plt.show()
```

```python
#  displot - ECDF (Empirical Cumulative Distribution Function)
sns.displot(
    data=iris,
    x='petal_length',
    hue='species',
    kind='ecdf', # Key variation: ECDF plot
    height=5
).fig.suptitle(' Distribution Plot (ECDF of Petal Length)', y=1.02)
plt.show()
```

```python
# jointplot - Joint and marginal distribution with a regression line
sns.jointplot(
    x='petal_length', y='petal_width',
    data=iris,
    kind='reg', # Key variation: Adds a linear regression line
)
plt.suptitle('Joint Plot (Scatter with Marginal Distributions & Regression)', y=1.02)
plt.show()
```

```python
# violinplot - Distribution density per species
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='sepal_width', data=iris, inner='quartile', palette="pastel") # inner='quartile' is a variation
plt.title('Violin Plot (Sepal Width Density by Species)')
plt.show()
```

```python
# scatterplot - Relationship between two features, colored by species
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, style='species', s=80)
plt.title(' Scatter Plot (Relating Sepal Features by Species)')
plt.show()
```

```python
#  countplot - Count of observations in each categorical bin
plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=iris, palette="coolwarm")
plt.title(' Count Plot (Number of Samples per Species)')
plt.show()
```

```python
# clustermap - Hierarchically-clustered heatmap
# This uses ALL numeric columns and reorders the rows/columns based on similarity
sns.clustermap(iris.drop('species', axis=1), cmap="mako", figsize=(8, 8))
plt.suptitle('Clustermap (Hierarchical Clustering of Features)', y=1.02)
plt.show()
```

```python
# lmplot - Scatter plot with a linear model fit (Regression visualization)
# Focuses on a single species (Iris-virginica)
sns.lmplot(
    x='petal_length', y='petal_width',
    data=iris.query("species == 'virginica'"), # Data subsetting
    height=6, ci=95, scatter_kws={'s': 50}
).fig.suptitle('Linear Model Plot (Relationship for Iris-virginica)', y=1.02)
plt.show()
```

```python
#  stripplot - Scatter plot where one variable is categorical
plt.figure(figsize=(8, 6))
sns.stripplot(x="species", y="petal_length", data=iris, jitter=True) # jitter=True prevents all points from overlapping
plt.title('Strip Plot (Petal Length Jittered by Species)')
plt.show()
```
