#1. Write a Python program which accepts the radius of a circle from the user and compute the area.
Sample input: Enter the radius: 3
Sample output: Area of Circle is 28.59999999999

```python
import math

def calculate_area(radius):
    return math.pi * radius**2

print("Circle Area Calculator")
radius_input = float(input("Enter the radius of the circle: "))

if radius_input < 0:
    print("Error: Radius cannot be negative.")
else:
    area = calculate_area(radius_input)
    print(f"The area of the circle with radius {radius_input} is: {area:.4f}")
```

#3. Write a Python program to get the largest number from a list.
Sample input: 12, 3, 47, 10
Sample output: 47 is the largest number.

```python
def find_largest_number(number_list):
    if not number_list:
        return None

    max_num = number_list[0]

    for number in number_list:
        if number > max_num:
            max_num = number

    return max_num

print("Largest Number Finder")

input_str = input("Enter a list of numbers : ")

number_list = [float(n.strip()) for n in input_str.split(',')]

largest = find_largest_number(number_list)

if largest is not None:
    print("The largest number is:", largest)
else:
    print("The list is empty.")
```

#4. Given a range of first 10 numbers, write a Python program to iterate from start number to the end number and print the sum of the current number and previous number.
Sample input: 1….10
Sample output: Current Number 1 Previous Number 0 Sum: 1 Current Number 2 Previous Number 1 Sum: 3 Current Number 3 Previous Number 2 Sum: 5… Current Number 10 Previous Number 9 Sum: 19

```python
def calculate_cumulative_sum(end_number):
    previous_number = 0

    print("Iteration | Current | Previous | Sum")
    print("-" * 35)

    for current_number in range(1, end_number + 1):
        current_sum = current_number + previous_number

        print(f"    {current_number:2}    |   {current_number:3}   |    {previous_number:3}   | {current_sum:3}")

        previous_number = current_number

print("Cumulative Sum of Sequence (1 to 10)")
calculate_cumulative_sum(10)
```

#5. Write a Python program to print only those numbers which are divisible of 5.
Sample input: 10, 20,33,46,55 Sample output: 10, 20, 55

```python
def divisible_by_five(number_list):
    divisible_list = []
    for number in number_list:
        if number % 5 == 0:
            divisible_list.append(number)
    return divisible_list

print("Divisible by five")

input_str = input("Enter a list of numbers : ")


number_list = [int(n.strip()) for n in input_str.split(',')]

result = divisible_by_five(number_list)

if result:
    print("\nNumbers divisible by 5:", result)
else:
    print("\nNo numbers divisible by 5 found in the list.")
```

#6. Write a Python program to check whether a number is prime or not. Sample input: Enter the number: 3
Sample output: 3 is a prime number.

```python
import math

def is_prime(n):
    if n <= 1:
        return False

    if n == 2:
        return True

    if n % 2 == 0:
        return False

    limit = math.isqrt(n)

    for i in range(3, limit + 1, 2):
        if n % i == 0:
            return False

    return True

number = int(input("Enter a whole number to check for prime: "))

if is_prime(number):
    print(f"\n{number} is a prime number.")
else:
    print(f"\n{number} is NOT a prime number.")
```

#7. Write a Python program to reverse a list using for loop.
Sample input: 10,40,30,70
Sample output: 70,30,40,10

```python
def reverse_list(input_list):
    n = len(input_list)

    for i in range(n // 2):
        temp = input_list[i]
        input_list[i] = input_list[n - 1 - i]
        input_list[n - 1 - i] = temp

    return input_list

print("List Reverser")

input_str = input("Enter a list of elements : ")

input_list = [item.strip() for item in input_str.split(',')]

original_list = list(input_list)

reversed_list = reverse_list(input_list)

print(f"Original list: {original_list}")
print(f"Reversed list: {reversed_list}")
```

#9. Write a Python program to find the maximum of three numbers
Sample input: 34,12,7
Sample output: 34

```python
def find_maximum(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

print("Enter three numbers to find the maximum.")
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

maximum = find_maximum(num1, num2, num3)

print("output:", int(maximum) if maximum == int(maximum) else maximum)
```
