# Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
def range_finder(l1):
  n=l1[0]             #setting the first value of the list as minimum value
  m=l1[0]             #setting the firts value of the list as maximun value
  for i in l1:        #looping through the entire list to check for maximum and minimum value.
    if i>=m :
      m=i
    if i<=n :
      n=i
  return (m,n)


lis1=input('Enter list of number: ')
list1 = [float(item.strip()) for item in lis1.split(',')]
print(list1)
maxi,mini=range_finder(list1)
print('Maximum : ',maxi)
print('Minimum : ',mini)
```

## Q2. Write function that returns mean value of a list of numbers

```python
def mean_finder(l1=[]):
  s=sum(l1)
  l=len(l1)
  mean=s/l
  return mean


lis1=input('Enter list of number: ')
list1 = [float(item.strip()) for item in lis1.split(',')]

print(list1)
mean1=mean_finder(list1)
print('Mean value = ',mean1)
```

# Q3. Write a function that returns the median value of a list of numbers
