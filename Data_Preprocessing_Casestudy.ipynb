```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#Loading Dataset

```python
hp=pd.read_csv('/content/House_Pricing.csv')
hp.head()
```

```python
hp.info()
```

```python
hp.describe()
```

#Duplicates Removal

```python
duplicate_rows = hp[hp.duplicated()]
duplicate_rows.shape
```

```python
duplicate_cols = hp.T.duplicated()
print(hp.columns[duplicate_cols])
```

#Missing Values Handling

```python
hp.isna().sum()
```

```python
hp.drop(columns=['No of Times Visited'],inplace=True,axis=1)
```

```python
hpp=hp.select_dtypes(include='number')
hpp.hist(figsize=(16,20),bins=50,xlabelsize=8,ylabelsize=8)
```

```python
hp.dropna(subset=['Zipcode','Latitude','Longitude','Living Area after Renovation (in Sqft)'],inplace=True)
```

```python
cols = ['Sale Price','No of Bathrooms','Flat Area (in Sqft)','Lot Area (in Sqft)',
        'Area of the House from Basement (in Sqft)'
        ]

for col in cols:
    hp[col] = hp[col].fillna(hp[col].median())

hp.isna().sum()
```

#Outliers Removal

```python
num_cols = hp.select_dtypes(include=['int64', 'float64']).columns


for col in num_cols:
    Q1 = hp[col].quantile(0.25)
    Q3 = hp[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    count = ((hp[col] < lower) | (hp[col] > upper)).sum()

    if count > 0:
        print(col, "=", count, "outliers")
        print(f"  Lower Bound: {lower}")
        print(f"  Upper Bound: {upper}")

        hp[col] = hp[col].clip(lower=lower, upper=upper)
```

#Encoding

```python
catg_cols = hp.select_dtypes(include='object').columns
hp= pd.get_dummies(hp, columns=catg_cols, drop_first=True)
hp.info()
```

#Scaling

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
hp[num_cols] = ss.fit_transform(hp[num_cols])
```

#Train-Test Split

```python
x = hp.drop('Sale Price', axis=1)
y = hp['Sale Price']


from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```
