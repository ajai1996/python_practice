#2.Create a Python program to calculate the factorial of a given number using a function.

```python
def fact(n):

  if n < 0:
        return "Factorial does not exist for negative numbers."
  if n == 0 or n == 1:
        return 1

  return n*fact(n - 1)


num = int(input("Enter a number: "))
print(f"The factorial of {num} is: {fact(num)}")
```

#3.Develop a Python program that checks whether a given word or phrase is a palindrome.

```python
def is_palindrome(word):

    word = word.lower()
    return word == word[::-1]
w=input("Enter a word or phrase: ")
print(is_palindrome(w))
```

#4.Create a Python program that takes a sentence as input and counts the number of vowels and consonants

```python
def count_vow_cons(word):

    vowels = {'a', 'e', 'i', 'o', 'u'}

    vowel = 0
    consonant= 0

    word = word.lower()

    for char in word:


            if char in vowels:
                vowel += 1
            else:
                consonant+= 1

    return vowel, consonant

s=input("Enter a sentence: ")

vowels,consonants = count_vow_cons(s)

print(f"Total Vowels:     {vowels}")
print(f"Total Consonants: {consonants}")
```

#5.Develop a Python program that takes two lists as input and returns a new list containing the common elements between them.

```python
def common(list1, list2):
    result = []
    for item in list1:

        if item in list2 and item not in result:
            result.append(item)
    return result

t1=input("Enter first list items : ").split()
t2=input("Enter second list items : ").split()

print(f"Common: {common(t1,t2)}")
```

#6.Write a Python program that takes a string as input and counts the frequency of each character in the string.

```python
from collections import Counter

def count_freq(text):

    return Counter(text)

input1 = input("Enter a string: ")

freq= count_freq(input1)

print("\n Character Frequencies : ")
for char, count in freq.most_common():

    print(f"{repr(char)}: {count}")
```

#7.Develop a Python program that checks if a given year is a leap year.

```python
def is_leap(year):
  if year % 4 == 0:

        if year % 100 == 0:
            if year % 400 == 0:
                return True
            else:
                return False

        else:
            return True

  else:
        return False

year = int(input("Enter a year : "))

if is_leap(year):
  print(f"\n {year} is a leap year.")
else:
  print(f"\n {year} is NOT a leap year.")
```

#8.Write a Python program that checks if a given number, is a prime number.

```python
import math

def is_prime(n):


    if n <= 1:
        return False

    if n == 2:
        return True


    if n % 2 == 0:
        return False


    limit = math.isqrt(n)


    for i in range(3, limit + 1, 2):
        if n % i == 0:

            return False

    return True


number = int(input("Enter a number to check for prime: "))

if is_prime(number):
        print(f"\n {number} is a prime number.")
else:
        print(f"\n {number} is NOT a prime number.")
```

#9.Write a Python program to generate the Fibonacci sequence up to a specified number of terms.

```python
def generate_fibonacci(n_terms):

    if n_terms <= 0:
        return []
    if n_terms == 1:
        return [0]

    sequence = [0, 1]

    while len(sequence) < n_terms:

        next_term = sequence[-1] + sequence[-2]
        sequence.append(next_term)

    return sequence


num_terms = int(input("Enter the number of Fibonacci terms to generate: "))
if num_terms < 0:
        print("Please enter a non-negative integer.")
else:
        fib_sequence = generate_fibonacci(num_terms)
        print(f"\nFibonacci Sequence (up to {num_terms} terms):")
        print(fib_sequence)
```

#10.Write a python program to find the roots of a quadratic equation

```python
import cmath
import math

def find_quadratic_roots(a, b, c):
    if a == 0:
        if b == 0:
            return "Error: This is not an equation (a=0, b=0).", None
        else:
            root = -c / b
            return f"This is a linear equation. Root: {root}", None

    discriminant = (b**2) - (4 * a * c)

    if discriminant > 0:
        sqrt_discriminant = math.sqrt(discriminant)
        root1 = (-b + sqrt_discriminant) / (2 * a)
        root2 = (-b - sqrt_discriminant) / (2 * a)
        return "Two distinct real roots:", (root1, root2)

    elif discriminant == 0:
        root = -b / (2 * a)
        return "One real (repeated) root:", (root, root)

    else:
        sqrt_discriminant = cmath.sqrt(discriminant)
        root1 = (-b + sqrt_discriminant) / (2 * a)
        root2 = (-b - sqrt_discriminant) / (2 * a)
        return "Two complex roots:", (root1, root2)

print("Welcome to the Quadratic Root Solver (ax^2 + bx + c = 0)")

a = float(input("Enter coefficient 'a': "))
b = float(input("Enter coefficient 'b': "))
c = float(input("Enter coefficient 'c': "))

nature, roots = find_quadratic_roots(a, b, c)

print("\n--- Results ---")
print(f"Equation: {a}x^2 + {b}x + {c} = 0")
print(f"Nature of Roots: {nature}")

if roots:
    root1_str = f"{roots[0].real:.4f}" if roots[0].imag == 0 else f"{roots[0]:.4f}"
    root2_str = f"{roots[1].real:.4f}" if roots[1].imag == 0 else f"{roots[1]:.4f}"

    print(f"Root 1 (x1): {root1_str}")
    print(f"Root 2 (x2): {root2_str}")
```
